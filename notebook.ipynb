```python
import json
import os.path
import urllib.request
```

```python
from gensim import matutils
from gensim.models.keyedvectors import KeyedVectors
import numpy as np
```

```python
EMOJI_URL = 'https://raw.githubusercontent.com/muan/emojilib/master/emojis.json'
EMOJI_NAME = 'emojilib.json'
BIN_NAME = 'GoogleNews-vectors-negative300.bin'

# Download emojilib
if not os.path.isfile(EMOJI_NAME):
    with open(EMOJI_NAME, 'wb') as f:
        with urllib.request.urlopen(EMOJI_URL) as response:
            f.write(response.read())
    print('Emojilib downloaded!')
else:
    print('Emojilib already downloaded')

from collections import defaultdict

with open('emojilib.json', 'r', encoding='utf-8') as f:
    emojis = json.load(f)

mapping = defaultdict(list)

for emoji_name, emoji in emojis.items():
    mapping[emoji_name].append(emoji['char'])
    for part in emoji_name.split('_'):
        mapping[part].append(emoji['char'])
    for keyword in emoji['keywords']:
        mapping[keyword].append(emoji['char'])
```

```python
mapping_list = list(mapping.items())
mapping_list.sort(key=lambda x: len(x[1]), reverse=True)
for name, emoji in mapping_list[:10] + mapping_list[-10:]:
    print('{}: {}'.format(name, ' '.join(emoji)[:80]))
```

```python
print('Loading model')
model = KeyedVectors.load_word2vec_format(BIN_NAME, binary=True)
print('Model loaded!')
```

```python
# Saving for efficiency
with open('vocab.json', 'w') as f:
    json.dump(model.index2word, f)
model.init_sims(replace=True)
model_norms = model.wv.vectors_norm
print(model_norms.shape)
norms = np.memmap('normsmemmap.dat', dtype=model_norms.dtype, mode='w+', shape=model_norms.shape)
norms[:] = model.syn0norm
del model, model_norms
```

```python
# Load saved
with open('vocab.json', 'r') as f:
    vocab = json.load(f)
norms = np.memmap('normsmemmap.dat', dtype=np.float32).reshape((-1, 300))
```

```python
lowercasedvocab = [word.lower() for word in vocab]
vocab_set = set(vocab) # For faster searching
lower_vocab_set = set(lowercasedvocab)
for word in list(mapping.keys()):
    if word not in vocab_set:
        if word.lower() in lower_vocab_set:
            cased_word = vocab[lowercasedvocab.index(word)]
            print('{} -> {}'.format(word, cased_word))
            mapping[cased_word] = mapping[word]
        else:
            print('delete ' + word)
        del mapping[word]
```

```python
vocab_norms = np.array([norms[vocab.index(word)] for word in mapping])
```

```python
dp = np.memmap('dpmap.dat', dtype=norms.dtype, mode='w+', shape=(norms.shape[0],))
CHUNKSIZE = 1000
for s in range(0, norms.shape[0], CHUNKSIZE):
    e = s+CHUNKSIZE
    dp[s:e] = np.amax(np.inner(norms[s:e], vocab_norms), axis=1)
```

```python
off = np.argpartition(dp, -100)[-100:]
sortd = off[np.argsort(dp[off])][::-1]
for i in sortd:
    print(vocab[i], dp[i])
```

```python
good_indices = np.where(dp > 0.5)
print(len(good_indices[0]))
good_vocab = np.array(vocab)[good_indices].tolist()
with open('good_vocab.txt', 'w', encoding='utf-8') as f:
    for word in good_vocab:
        f.write(word + ' ')
good_weights = norms[good_indices]
np.save('good_weights.npy', good_weights)
```

```python
with open('good_vocab.txt', encoding='utf-8') as f:
    good_vocab = f.read().split(' ')[:-1]
with open('mapping.json', encoding='utf-8') as f:
    mapping = json.load(f)
good_weights = np.load('good_weights.npy')
```

```python
mapping = dict(mapping)
lookup = lambda word: good_weights[good_vocab.index(word)]
emoji_words = list(mapping.keys())
emoji_norms = np.array([lookup(word) for word in emoji_words])

inp = lookup('ecstatic')
dot = np.dot(emoji_norms, inp)
matches = np.argpartition(dot, -10)[-10:]
sortedmatches = matches[np.argsort(dot[matches])][::-1]
for index in sortedmatches:
    print(dot[index], emoji_words[index], ' '.join(mapping[emoji_words[index]]))
```

```python
from itertools import permutations 

with open('data/emoji.txt', 'w') as f:
    f.write(' '.join(emoji_words))
np.save('data/emoji.npy', emoji_norms)

firsts = np.array([v[0].lower() for v in good_vocab])
seconds = np.array([v[1].lower() if len(v) > 1 else ' ' for v in good_vocab])
thirds = np.array([v[2].lower() if len(v) > 2 else ' ' for v in good_vocab])

possible = sorted(set(firsts) | set(seconds) | set(thirds))
allowed = possible[:possible.index('~')+1]
print(allowed)

fs_forbid = '<>:"/\\|?*' # Characters forbidden by file system

# initial = set()
# for word in good_vocab:
#     #if word.startswith('1'):
#     initial.add(word[0].lower())
# print(sorted(list(initial)))

def save(letters, indices):
    filename = ''.join(
        chr(ord(c)+128) if c in fs_forbid else c for c in letters
    )
    with open('data/dat-{}.txt'.format(filename), 'w', encoding='utf-8') as f:
        f.write(' '.join(good_vocab[i] for i in indices[0]))
    np.save('data/dat-{}.npy'.format(filename), good_weights[indices])

threefiles = []
for fl, sl in permutations(allowed, 2):
    if sl == ' ':
        print(fl)
    matched = np.logical_and(firsts == fl, seconds == sl)
    if np.count_nonzero(matched) == 0:
        continue
    elif np.count_nonzero(matched) < 1000:
        save(fl + sl, np.where(matched))
        continue
    threefiles.append(fl + sl)
    for tl in allowed:
        indices = np.where(np.logical_and(matched, thirds == tl))
        if indices[0].size == 0:
            continue
        save(fl + sl + tl, indices)
```
